```
import requests
import cv2
import os
 
from imutils import paths
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/nupursjsu/Deep-Learning/master/homework_1/donald_duck.csv', sep='delimiter', header=None)
data=data[0]
data.to_csv('download',index=False)

url_path = open('download').read().strip().split('\n')
total = 0
print(url_path)
if not os.path.exists('donald_duck'):
    os.mkdir('donald_duck')
image_path = 'donald_duck'
 
for url in url_path:
    try:
        req = requests.get(url, timeout=60)
 
        file_path = os.path.sep.join([image_path, '{}.jpg'.format(
            str(total).zfill(6))]
        )
        file = open(file_path, 'wb')
        file.write(req.content)
        file.close()
 
        print('Downloaded {}'.format(file_path))
        total += 1
 
    except:
        print('Could not download {}. Downloading next file'.format(file_path))
```

```
import requests
import cv2
import os
 
from imutils import paths
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/nupursjsu/Deep-Learning/master/homework_1/mickey_mouse.csv', sep='delimiter', header=None)
data=data[0]
data.to_csv('download',index=False)

url_path = open('download').read().strip().split('\n')
total = 0
print(url_path)
if not os.path.exists('mickey_mouse'):
    os.mkdir('mickey_mouse')
image_path = 'mickey_mouse'
 
for url in url_path:
    try:
        req = requests.get(url, timeout=60)
 
        file_path = os.path.sep.join([image_path, '{}.jpg'.format(
            str(total).zfill(6))]
        )
        file = open(file_path, 'wb')
        file.write(req.content)
        file.close()
 
        print('Downloaded {}'.format(file_path))
        total += 1
 
    except:
        print('Could not download {}. Downloading next file'.format(file_path))
```

```
path='images'
data = (ImageList.from_folder(path) .random_split_by_pct() .label_from_folder() .transform(get_transforms(), size=224) .databunch())
```

```
data.show_batch(rows=3, figsize=(7,6))
```

```
learn = cnn_learner(data, models.resnet34, metrics=error_rate)
```

```
learn.model
```

```
learn.fit_one_cycle(4)
```

```
learn.save('stage-1')
```

```
pred = ClassificationInterpretation.from_learner(learn)
```

```
losses,idxs = pred.top_losses()

len(data.valid_ds)==len(losses)==len(idxs)
```

```
pred.plot_top_losses(9, figsize=(15,11))
```

```
pred.plot_confusion_matrix(figsize=(12,12), dpi=60)
```

```
pred.most_confused(min_val=2)
```
